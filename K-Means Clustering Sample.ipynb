```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
data[1]#Cluster it belongs to
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4)
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_ #Centroids of the cluster
```

```python
km.labels_ #Labels it believes to be true for the cluster (Instead of Predict in Supervised)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=km.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
